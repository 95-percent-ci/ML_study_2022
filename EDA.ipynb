```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size=0.30, random_state=343434)
```

## OLS FORM

```python
# Assuming actual is of this form #
```

$y = mx + c + \epsilon$

```python
# y = deterministic + random_part # 
```

```python
# Where `e` is random error (Totally unpredictable)
```

```python
# Estimate of y is written as : which is also estimate of deterministic part #
```

$\hat{y} = \hat{m}x + \hat{c}$

Solving for $\hat{m}$ and $\hat{c}$

```python
x_train, y_train = df_train['height'].values, df_train['weight'].values
```

```python
x_test, y_test = df_test['height'].values, df_test['weight'].values
```

```python
x_train = x_train.reshape(-1,1)
```

```python
x_test = x_test.reshape(-1,1)
```

```python
lm = LinearRegression()
```

```python
x_train[0:20]
```

```python
lm.fit(x_train[0:20], y_train[0:20])
```

```python
y_test_pred = lm.predict(x_test)
```

```python
print("Coefficients: \n", lm.coef_, lm.intercept_)
```

```python
print("Coefficients: \n", lm.coef_, lm.intercept_)
```

```python
r2_test = r2_score(y_test, y_test_pred)
```
